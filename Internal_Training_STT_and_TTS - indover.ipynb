## Speech Recognition

Biometrik merupakan salah satu metode atau cara untuk mengenali manusia berdasarkan satu atau lebih ciri-ciri fisik atau tingkah laku yang unik. Salah satu biometrik yang dapat digunakan adalah suara melalui pengenalan suara atau speech recognition. Suara atau secara luas kita sebut ucapan merupakan bentuk komunikasi paling alami bagi manusia. 

Speech Recognition merupakan sistem yang digunakan untuk mengenali perintah kata dari suara manusia dan kemudian diterjemahkan menjadi suatu data yang dimengerti oleh komputer. Inputan berupa suara yang ditangkap oleh perangkat audio yang akan diterjemahkan/dikonversi dari sinyal akustik menjadi sebuah keluaran berupa teks. Pada saat ini, sistem ini digunakan untuk menggantikan peranan input dari keyboard ataupun mouse. Kata-kata yang ditangkap dan dikenali bisa jadi sebagai hasil akhir untuk sebuah aplikasi seperti command & control, penginputan data, dan persiapan dokumen. 

### Ide dasar Speech Recognition

Komunikasi antar manusia tidak akan terjadi tanpa sebuah ucapan. Saat berbicara kita melakukan sintesis ucapan, pengenalan ucapan, dan pemahaman ucapan. Komunikasi melalui sebuah ucapan dapat dirumuskan dan digambarkan sebagai berikut:

Tujuan utama dari Automation Speech Recognition (ASR) adalah membangun sistem yang dapat mensimulasikan bagaimana cara mendengar manusia, yaitu ia dapat memahami bahasa yang diucapkan dan merespons dengan kata lain bahwa sistem dapat bereaksi dengan tepat terhadap kata-kata yang diucapkan dan mengonversi ucapan ke media lain seperti teks.

### Jenis-jenis pengenalan ucapan

Berdasarkan kemampuan dalam mengenal kata yang diucapkan, terdapat 5 jenis pengenalan kata, yaitu:
- Isolated Words:

Proses pengidentifikasian kata yang hanya dapat mengenal kata yang diucapkan jika kata tersebut memiliki jeda waktu pengucapan antar kata

- Connected Words:

Proses pengidentifikasian kata yang mirip dengan kata-kata terisolasi, namun membutuhkan jeda waktu pengucapan antar kata yang lebih singkat

- Continuous Speech: 

Proses pengidentifikasian kata yang sudah lebih maju karena dapat mengenal kata-kata yang diucapkan secara berkesinambungan dengan jeda waktu yang sangat sedikit atau tanpa jeda waktu. Proses pengenalan suara ini sangat rumit karena membutuhkan metode khusus untuk membedakan kata-kata yang diucapkan tanpa jeda waktu. Pengguna perangkat ini dapat mengucapkan kata-kata secara natural

- Spontaneous Speech:

Proses pengidentifikasian kata yang dapat mengenal kata-kata yang diucapkan secara spontan tanpa jeda waktu antar kata

- Voice Verification/Identification:

Proses pengidentifikasian kata yang tidak hanya mampu mengenal kata, namun juga mengidentifikasi siapa yang berbicara.

### Proses pengenalan suara/ujaran

1. Skema Utama Speech Recognition

   Komponen utama dalam sistem pengenalan suara :

    - Penerimaan data input : Ketika kita berbicara, kita menghasilkan getaran di udara. Getaran tersebut membentuk sebuah siynal yang selanjutnya kita sebut sebagai sinyal analog. Suara yang kita ucapkan kemudian diterima oleh device input yang berupa mikrofon. Mikrofon inilah yang selanjutnya menangkap signal analog kemudian meneruskannya melalui sinyal listrik dan mengubahnya menjadi sinyal digital yang dapat dibaca oleh komputer. Untuk dapat mengubah sinyal dari analog ke digital dibutuhkan sebuah converter yang disebut dengan Analog to Digital Converter(ADC).ADC akan mengambil sample dari suara kita yang berupa gelombang pengukuran suara secara tepat dan berkala. 
    - Ekstraksi fitur (feature extraction): proses mengidentifikasi bagian-bagian dari input yang mengandung ucapan dan kemudian mengubah bagian-bagian itu menjadi sebuah urutan-urutan atau yang dikenal sebagai parameter akustik.
    - Data / korpus untuk pelatihan dan pengujian: Basis data ini adalah kumpulan ucapan yang direkam. Korpus ini harus cukup besar dan cukup relevan untuk mencakup keragaman bicara.
    - Pemodelan akustik: Model akustik mengambil bentuk gelombang bicara dan memecahnya menjadi fragmen kecil dan memprediksi fonem yang paling mungkin dalam sebuah ucapan.
    - Pemodelan pelafalan: Model pelafalan mengeluarkan suara dan mengikatnya bersama untuk membentuk kata(mengaitkan kata dengan representasi fonetisnya)
    - Pemodelan bahasa: Model bahasa mengambil kata-kata dan mengikatnya bersama-sama untuk membentuk kalimat, yaitu memprediksi urutan kata yang paling mungkin (atau string teks) di antara beberapa set string teks.
    - Decoder : menggabungkan prediksi model akustik dan bahasa dan menghasilkan string teks yang paling mungkin untuk input file pidato yang diberikan

2. Tahap Blok Speech Recognition dengan HMM

    - Tahap feature extraction: Pemfilteran sinyal suara dan mengubah sinyal suara analog ke digital. Ekstraksi fitur digunakan dalam fase pelatihan dan pengenalan. Tujuan utama Ekstraksi Fitur adalah untuk menyederhanakan pengenalan dengan meringkas sejumlah besar data ucapan tanpa kehilangan sifat akustik yang mendefinisikan ucapan tersebut. Ekstraksi fitur terdiri dari langkah-langkah berikut:
        - Frame Blocking : Sebuah investigasi menunjukkan bahwa karakteristik sinyal wicara tetap diam dalam periode waktu yang cukup singkat (quasi-stationary). Untuk alasan tersebut, sinyal ucapan diproses dalam interval waktu singkat dan terbagi menjadi bingkai(frame) dengan ukuran umumnya antara 30 dan 100 miliseconds. Setiap frame tumpang tindih dengan frame sebelumnya dengan ukuran yang telah ditentukan. Tujuan dari skema yang tumpang tindih adalah untuk memperlancar transisi dari frame ke frame.
        - Windowing : Langkah kedua adalah windowing. Hal ini dilakukan untuk menghilangkan diskontinuitas pada tepi bingkai. Jika fungsi windowing didefinisikan sebagai w (n), 0 <n <N-1 di mana N adalah jumlah sampel di setiap frame, sinyal yang dihasilkan adalah y (n) = x (n) w (n). Metode umum yang sering digunakan adalah Hamming Windows.  
        - FFT (Fast Fourier Transform): Langkah selanjutnya adalah mengambil Fast Fourier Transform dari setiap frame. Transformasi ini mengubah domain dari waktu ke frekuensi.
        - Mel-Frequency Wrapping : Sinyal untuk setiap frame dilewatkan melalui filter pass band Mel-Scale untuk meniru telinga manusia
        - Cepstrum (Mel Frequency Cepstral Coefficients): Pada langkah terakhir, dilakukan inverse terhadap frame Fourier untuk membawa mereka kembali ke domain waktu menggunakan Discrete  Cosine Transform yang dirumuskan dengan:

dimana:

   - Tahap tugas pemodelan: Untuk pemodelan dibuatkan suatu model HMM dari data-data yang berupa sample ucapan dari sebuah kata (berupa data digital). Setelah dilakukan segmentasi, bagian-bagian dari matriks MFCC yang sesuai dengan masing-masing fonem digunakan dalam konstruksi model HMM. Misalnya, dari MFCC kata "mereka", kita mendapatkan satu model untuk "m", "r", "k", "a", dan dua model untuk "e". Selama melatih beberapa kata, ini memungkinkan untuk membuat model fonem yang sama secara berulang. Pada kasus ini, model fonem diberi nama seperti “_e_1”, “_e_2”, “_e_15”, dan seterusnya.
   
   - Tahap sistem pengenalan HMM: Pada tahap ini, ucapan yang tidak dikenal harus diidentifikasi sebagai kata dalam bahasa Indonesia (contoh kasus adalah kata "mereka") menggunakan model fonem yang diperoleh pada tahap penerimaan ucapan. Tahap pengenalan ini dapat diartikan sebagai proses pelatihan. 

3. Proses Pencocokan Pola Suara
    Karena pelatihan dilakukan pada tingkat fonem, maka pengenalan model HMM juga harus dilakukan pada level yang sama. Saat melakukan pelatihan, panjang fonem terkecil dari semua fonem disimpan dalam sebuah parameter. Pada tahap pencocokan ini, ucapan yang tidak dikenal dibagi menjadi bagian-bagian yang lebih kecil dari nilai parameter yang tersimpan agar tidak setiap fonemnya dapat dikenali tanpa terlewat satupun. Panjang pemisahan minimum paling optimal yang ditemukan oleh sebuah eksperimen adalah sekitar setengah dari nilai parameter yang disimpan. 
    
    Setelah dilakukan pemisahan pada setiap bagian dari ucapan, pencarian HMM Viterbi dibuat dengan vektor fitur yang diekstraksi dari bagian itu. 


    
   Secara umum, speech recognition memproses sinyal suara yang masuk dan menyimpannya dalam bentuk digital. Hasil proses digitalisasi tersebut kemudian dikonversi dalam bentuk spektrum suara yang akan dianalisa dengan membandingkannya dengan template suara pada database sistem. ### Contoh Speech to Text Sederhana

```python
#Library yang dibutuhkan untuk menjalankan Speech to Text
import speech_recognition as sr
import pyaudio
import sys, os
```

```python
r = sr.Recognizer()
with sr.Microphone() as source:
    r.pause_threshold = 1
    # listen for 1 second to calibrate the energy threshold for ambient noise levels
    r.adjust_for_ambient_noise(source)
    while True:
        print("Tolong katakan sesuatu!")
        audio = r.listen(source)
        try:
            get_text = r.recognize_google(audio, language = 'id-ID').lower()
            print('Kamu bilang: ' + get_text + '\n')
            if(get_text == "keluar"):
                print("Oke, sampai ketemu lagi bosku!")
                sys.exit()
        except sr.UnknownValueError:
            print("Aku tidak paham dengan yang kamu katakan!\n")
```

```python
# Import library Speech Recognition dan pendukungnya.
import speech_recognition as sr
import pyaudio
import sys, os
from gtts import gTTS
import subprocess
import datetime
```

```python
# Ambil Bulan
bulan = ["Januari","Februari","Maret","April","Mei","Juni","Juli","Agustus","September","Oktober","November","Desember"]

# Buat ke dalam fungsi yang akan membaca text
def speech_out(text):
    tts = gTTS(text, lang='id')
    tts.save('audio.mp3')
    os.system("mpg123 audio.mp3")

speech_out("halo Tuanku apakah ada yang bisa dibantu?")
print("Mesin: Halo Tuanku, Apakah ada yang bisa dibantu?\n")

silent = 0
r = sr.Recognizer()
with sr.Microphone() as source:
    r.pause_threshold = 1
    r.adjust_for_ambient_noise(source)
    while True:
        audio = r.listen(source)
        try:
            get_text = r.recognize_google(audio, language = 'id-ID').lower()
            print('Kamu bilang: ' + get_text + '\n')
            
            if("buka" in get_text):
                filename = get_text.split()[get_text.split().index("buka")+1]
                
                if(os.system("start "+filename+".exe")!=0):
                    speech_out("Maaf,"+ filename + " tidak ditemukan")
                    print("Mesin: Maaf,"+ filename + " tidak ditemukan\n")
                else:
                    speech_out("Oke")
                    print("Mesin: Oke\n")
            
            if("tutup" in get_text):
                filename = get_text.split()[get_text.split().index("tutup")+1]
                
                speech_out("Anda yakin ingin menutup "+filename)
                print("Mesin: Anda yakin ingin menutup " + filename + "?\n")
                
                #/im Menentukan nama proses yang akan dihentikan.
                #/f Menentukan bahwa proses diberhentikan secara paksa.
                if(os.system("taskkill /im "+filename+".exe /f")!=0):
                    speech_out("Maaf,"+ filename + " tidak ditemukan")
                    print("Mesin: Maaf,"+ filename + " tidak ditemukan\n")
                else:
                    speech_out("Oke")
                    print("Mesin: Oke\n")
                    
            if("jam" in get_text and "berapa" in get_text):
                speech_out(datetime.datetime.now().strftime("%H:%M:%S"))
                print("Mesin: pukul "+ datetime.datetime.now().strftime("%H:%M:%S") + "\n")
            
            if("tanggal" in get_text and "berapa" in get_text):
                tanggal = str(datetime.datetime.now().date()).split("-")
                speech_out(tanggal[2] + " " + bulan[int(tanggal[1])] + " " +tanggal[0])
                print("Mesin: tanggal "+ tanggal[2] + " " + bulan[int(tanggal[1])] + " " +tanggal[0] + "\n")
            
            if(get_text == "keluar" or get_text == "tidak" or get_text == "nggak"):
                speech_out("Ok, sampai ketemu lagi!")
                print("Mesin: Ok, sampai ketemu lagi!\n")
                sys.exit()
                
            speech_out("apakah ada lagi yang bisa dibantu?")
            print("Mesin: Apakah ada lagi yang bisa dibantu?\n")
        except sr.UnknownValueError:
            if(silent<3):
                speech_out("Maaf, aku tidak dengar suaramu. Coba ulangi lagi!")
                print("Mesin: Maaf, aku tidak dengar suaramu. Coba ulangi lagi!\n")
                silent += 1
            else:
                speech_out("Sepertinya kamu pergi, sampai ketemu lagi!")
                print("Mesin: Sepertinya kamu pergi, sampai ketemu lagi!\n")
                sys.exit()
```

## Kelebihan dan Kelemahan Speech Recognition

Speech Recognition memiliki beberapa kelebihan. Teknologi ini dapat mempercepat transmisi informasi dan umpan balik dari transmisi tersebut. Contohnya pada komando/perintah suara. Hanya dalam selang waktu sekitar satu atau dua detik setelah kita mengkomandokan perintah melalui suara, komputer sudah memberi umpan balik atas perintah suara kita. Selain itu, kemudahan lainnya adalah perintah yang kita berikan ke komputer dilakukan tanpa perangkat keras seperti mouse dan keyboard atau dengan kata lain kita hanya menggunakan suara kita saja untuk memasukkan perintah ke komputer sehingga hal ini sangat memudahkan pekerjaan.

Selain memiliki kelebihan, speech recognition juga memiliki kelemahan yaitu rawan terhadap gangguan atau noise. Hal ini disebabkan oleh proses sinyal suara yang masih berbasis frekuensi. Ketika sebuah informasi dalam sinyal suara mempunyai komponen frekuensi yang sama banyaknya dengan komponen frekuensi gangguannya, akan sulit untuk memisahkan gangguan dari sinyal suara. Selain itu, jumlah kata yang dapat dikenal juga terbatas. Hal ini disebabkan pengenal ucapan bekerja dengan cara mencari kemiripan dengan basis data yang dimiliki.
